## QLearning in Connect2, Connect3, Connect4

### Qlearning
Q-learning is an algorithm of TD class used in reinforcement learning. It is an-algorithm OFF-Policy where the learned action-value function, Q, directly approximates the optimal action-value function, independent of the policy being followed.

The goal of Q-learning is to learn what action an agent has to take under specific circumstances maximizing the expected value of the total reward per discount, starting from the current state choosing a certain action. It does not require a model of the environment and can handle problems with stochastic transitions and rewards, without requiring adaptations.

#### Epsilon-Greedy policy
Epsilon-Greedy is a policy that chooses to explore an action with probability ε and to exploit an known action with probability 1-ε.
1. Exploration: seeking an action better than those known choosing a random action.
2. Exploitation: exploiting the action known that gives maximum reward in given state.


#### Pseudo-code
1. alpha: determines to what extent newly acquired information overrides old information
2. gamma: determines the importance of future rewards.
3. epsilon: determine the probability of exploration rather than exploitation

Note: 0 < α , γ , ε < 1


#### Trace
The trace is a memory buffer containing traces of past events (visited states, actions ...). The trace evaporates over time because is always updating. It allows you to give a delayed reward over time.


### The project

The project uses QLearning algorithm with epsilon-greedy policy and a trace. Two agents learn to play ConnectX by playing against. An agent learns how to play first always playing first and other one learns how to play second always playing second.
Six agents will be created, two for each mode of Connect: aget1_connect2 and agent2_connect2, agent1_connect3 and agent2_connect3, agent1_connect4 and agent2_connect4.

By choice:
1. Agent1: plays first. Plays yellow coins.
2. Agent2: plays secnd. Plays red coins.
3. Board:
    1. Connect2: 7x6
    2. Connect3: 5x4
    3. Connect2: 4x2
   
There is a configuration file that can be edited to change default value.

#### Database creation
Before launching the python script it is necessary to create the database in which Qs will be stored. 
As follows, have to be created three database, one for each mode of connect, with the relative table:

DB : TABLE
1. connect2.db : connect3Qs 
2. connect3.db : connect3Qs
3. connect4.db : connect4Qs



#### How to lunch
Approximately with default alpha (0.3), gamma (0.4), epsilon (0.4):
1. connect2: at least ~15,000 games are needed to train the agent.
2. connect3: at least ~15,000 games are needed to train the agent.
3. connect4: at least ~15,000 games are needed to train the agent.

The default number of games is 10000. Don't worry you can change it at run time.

### Note
1. In order to reduce the space and time complexity in Connect3 and Connect4, which is very enormus (~10^13 in Connect4), the first action of the Agent1 is always insert the coin in the central column of the board, which is considered the most promising. Agent1 hasn't restrictions in Connect2 because space and time complexity is very reduced.
2. In singleplayer mode the player plays always second with red coins replacing Agent2. 
3. During the training the Agent2 stores the inverted state rather than state. Inverted state is the state where the color of the coins is inverted: if a coin in the state is red that coin is yellow in the inverted state, vice versa with yellow coins. In this way the project is easily extendible: 
    1. You could choose who starts first randomly. Stored the inverted state the player (red) will play against the Agent1 if he play second or against Agent2 otherwise.
    2. In addition to the first, you should give the Agent1 the possibility to insert a coin in any column while it chooses the first action. This is to avoid that the player play against a stupid agent. In fact this project assumes that the Agent plays always first and insert the first coin in the central column. So, the Agent2 in this project learn how to play second only if the first play in the first column.
